## 2. Загрузка модели
Загружаем файл `ml_model.pkl`, сформированный в результате работы Airflow DAG.

```python
import joblib
import pandas as pd
import os

# Путь к модели (учитываем структуру папок)
model_path = 'data/ml_model.pkl' if os.path.exists('data/ml_model.pkl') else 'ml_model.pkl'
model = joblib.load(model_path)
print("Модель успешно загружена.")
```

## 3. Прогнозирование
Для Варианта 18 (Стокгольм, 3 дня) средняя температура составила 5.5°C.

```python
my_temp = 5.5
input_data = pd.DataFrame({'temperature': [my_temp]})
prediction = model.predict(input_data)

print(f"При температуре {my_temp}°C прогнозируемые продажи составят: {prediction[0]:.2f}")
```
